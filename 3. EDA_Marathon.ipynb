## Comparing speeds Male VS Female, by 50km and 50mi races

```python
(df3
    .groupby(['race_length', 'athlete_gender'])
    ['athlete_average_speed']
    .mean()
 )
```

## Top 15 age groups in the 50mi race (at least 20 participants)

```python
(df3
    .query('race_length == "50mi"')
    .groupby('athlete_age')
    ['athlete_average_speed']
    .agg(['mean', 'count'])
    .sort_values('mean', ascending = False)
    .query('count>19')
    .head(15)
 )
```

## Worst 20 age groups in the 50mi race (at least 10 participants)

```python
(df3
    .query('race_length == "50mi"')
    .groupby('athlete_age')
    ['athlete_average_speed']
    .agg(['mean', 'count'])
    .sort_values('mean', ascending = True)
    .query('count>9')
    .head(20)
 )
```

## Comparing speeds by seasons

### Seasons and their respective months (Northern Hemisphere)
* Spring  3-5
* Summer  6-8
* Fall    9-11
* Winter 12-2

### Creating a new `race_month` column

```python
df3['race_month'] = (df3
                        ['race_day']
                        .str.split('.')
                        .str.get(1)
                        .astype(int)
                     )
```

```python
df3.head()
```

### Creating a new `race_season` column

```python
df3['race_season'] = (df3
                        ['race_month']
                        .apply(lambda x: 
                                'Winter' if x > 11 
                                else 'Fall' if x > 8 
                                else 'Summer' if x > 5 
                                else 'Spring' if x > 2 
                                else 'Winter'
                                )
                      )
```

```python
df3.head(25)
```

### Grouping by season: 50mi races

```python
(df3
    .query('race_length == "50mi"')
    .groupby('race_season')
    ['athlete_average_speed']
    .agg(['mean', 'count'])
    .sort_values('mean', ascending = False)
 )
```

### Grouping by season: 50km races

```python
(df3
    .query('race_length == "50km"')
    .groupby('race_season')
    ['athlete_average_speed']
    .agg(['mean', 'count'])
    .sort_values('mean', ascending = False)
 )
```

### Observation
It seems like it's tougher to run in Summer, especially for the 50mi race, as compared to the 50km race.
